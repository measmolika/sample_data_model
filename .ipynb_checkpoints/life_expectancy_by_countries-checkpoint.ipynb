```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
```

```python
data = pd.read_csv("life_expectancy_data.csv") #data src = "https://www.kaggle.com/kumarajarshi/life-expectancy-who"
```

```python
data.head()
```

```python
data.describe()
```

```python
countries = ["United States of America","Japan","China"]

for country in countries:
    arr = data[['year','life_expectancy']].loc[data['country']==country]
    plt.plot(arr.year,arr.life_expectancy,label=country)

plt.legend()
plt.title('Life expectancy by countries')
plt.xlabel('year')
plt.ylabel('life_expectancy')
sns.despine
```
